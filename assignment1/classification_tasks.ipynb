# Classification Tasks with Logistic Regression

## 1. Imports and Setup

```python
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions

from LogisticRegression import LogisticRegression

# Ensure the directory for saving model parameters exists
params_dir = "ModelParams"
os.makedirs(params_dir, exist_ok=True)
```

## 2. Data Loading

```python
iris = load_iris()
X = iris.data
y = iris.target
```

---
### Variant 1: Petal Length & Width
---

```python
# Prepare petal data
X_petal = iris.data[:, 2:4] # Petal length and width
y_petal = iris.target

# Split data
X_train_petal, X_test_petal, y_train_petal, y_test_petal = train_test_split(
    X_petal, y_petal, test_size=0.2, random_state=42, stratify=y_petal
)

# Scale features
scaler_petal = StandardScaler()
X_train_scaled_petal = scaler_petal.fit_transform(X_train_petal)
X_test_scaled_petal = scaler_petal.transform(X_test_petal)

# Train model
model_petal = LogisticRegression(max_epochs=500, patience=20, regularization=0.01)
model_petal.fit(X_train_scaled_petal, y_train_petal, learning_rate=0.01)

# Evaluate accuracy
y_pred_petal = model_petal.predict(X_test_scaled_petal)
accuracy_petal = np.mean(y_pred_petal == y_test_petal)
print(f"Accuracy for Logistic Regression (Petal Features) on test set: {accuracy_petal:.4f}")

# Save weights
model_petal_path = os.path.join(params_dir, "logistic_petal_params.npz")
model_petal.save(model_petal_path)

print("Petal model training complete and weights saved.")
```

```python
# Visualize Decision Regions for Petal model
plt.figure(figsize=(10, 6))
plot_decision_regions(X_train_scaled_petal, y_train_petal, clf=model_petal, legend=2)
plt.title('Logistic Regression Decision Regions (Petal Features)')
plt.xlabel('Petal Length [scaled]')
plt.ylabel('Petal Width [scaled]')
plt.show()
```

---
### Variant 2: Sepal Length & Width
---

```python
# Prepare sepal data
X_sepal = iris.data[:, 0:2] # Sepal length and width
y_sepal = iris.target

# Split data
X_train_sepal, X_test_sepal, y_train_sepal, y_test_sepal = train_test_split(
    X_sepal, y_sepal, test_size=0.2, random_state=42, stratify=y_sepal
)

# Scale features
scaler_sepal = StandardScaler()
X_train_scaled_sepal = scaler_sepal.fit_transform(X_train_sepal)
X_test_scaled_sepal = scaler_sepal.transform(X_test_sepal)

# Train model
model_sepal = LogisticRegression(max_epochs=500, patience=20, regularization=0.01)
model_sepal.fit(X_train_scaled_sepal, y_train_sepal, learning_rate=0.01)

# Evaluate accuracy
y_pred_sepal = model_sepal.predict(X_test_scaled_sepal)
accuracy_sepal = np.mean(y_pred_sepal == y_test_sepal)
print(f"Accuracy for Logistic Regression (Sepal Features) on test set: {accuracy_sepal:.4f}")

# Save weights
model_sepal_path = os.path.join(params_dir, "logistic_sepal_params.npz")
model_sepal.save(model_sepal_path)

print("Sepal model training complete and weights saved.")
```

```python
# Visualize Decision Regions for Sepal model
plt.figure(figsize=(10, 6))
plot_decision_regions(X_train_scaled_sepal, y_train_sepal, clf=model_sepal, legend=2)
plt.title('Logistic Regression Decision Regions (Sepal Features)')
plt.xlabel('Sepal Length [scaled]')
plt.ylabel('Sepal Width [scaled]')
plt.show()
```

---
### Variant 3: All Features
---

```python
# Prepare full data
X_all = iris.data
y_all = iris.target

# Split data
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42, stratify=y_all
)

# Scale features
scaler_all = StandardScaler()
X_train_scaled_all = scaler_all.fit_transform(X_train_all)
X_test_scaled_all = scaler_all.transform(X_test_all)

# Train model
model_all = LogisticRegression(max_epochs=500, patience=20, regularization=0.01)
model_all.fit(X_train_scaled_all, y_train_all, learning_rate=0.01)

# Evaluate accuracy
y_pred_all = model_all.predict(X_test_scaled_all)
accuracy_all = np.mean(y_pred_all == y_test_all)
print(f"Accuracy for Logistic Regression (All Features) on test set: {accuracy_all:.4f}")

# Save weights
model_all_path = os.path.join(params_dir, "logistic_all_params.npz")
model_all.save(model_all_path)

print("All-features model training complete and weights saved.")
```
